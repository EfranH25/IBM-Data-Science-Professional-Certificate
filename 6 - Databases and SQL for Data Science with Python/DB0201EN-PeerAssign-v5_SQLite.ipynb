```python
def config(filename='database.ini', section='postgresql'):
    # create a parser
    parser = ConfigParser()
    # read config file
    parser.read(filename)

    # get section, default to postgresql
    db = {}
    if parser.has_section(section):
        params = parser.items(section)
        for param in params:
            db[param[0]] = param[1]
    else:
        raise Exception('Section {0} not found in the {1} file'.format(section, filename))

    return db

param = config()
param

conn_string = 'postgresql://{}:{}@{}:{}/{}'.format(param['user'],param['password'],param['host'],param['port'],param['database'])
```

##### List community areas with per capita income less than 11000.


```python
cursor.execute('SELECT "COMMUNITY_AREA_NAME", "PER_CAPITA_INCOME" from public."chicago_census_data" WHERE "PER_CAPITA_INCOME" < 11000 ORDER BY "PER_CAPITA_INCOME" DESC')

for i in cursor.fetchall():
    print(i)
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
cursor.execute('SELECT "CASE_NUMBER", "PRIMARY_TYPE", "DESCRIPTION" FROM public."chicago_crime_data" WHERE "DESCRIPTION" LIKE \'%MINOR%\'')

for i in cursor.fetchall():
    print(i)
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
cursor.execute('SELECT "CASE_NUMBER", "PRIMARY_TYPE", "DESCRIPTION" FROM public."chicago_crime_data" WHERE "DESCRIPTION" LIKE \'%CHILD%\' AND "PRIMARY_TYPE" = \'KIDNAPPING\'')

for i in cursor.fetchall():
    print(i)
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
cursor.execute('SELECT "PRIMARY_TYPE" P, "LOCATION_DESCRIPTION" L, COUNT(*) FROM public."chicago_crime_data" WHERE "LOCATION_DESCRIPTION" LIKE \'%SCHOOL%\' GROUP BY P, L')

for i in cursor.fetchall():
    print(i)
```

### Problem 6

##### List the average safety score for each type of school.


```python
cursor.execute('SELECT "Elementary, Middle, or High School" T, AVG("SAFETY_SCORE") A FROM public."chicago_school_data" GROUP BY T ORDER BY A')

for i in cursor.fetchall():
    print(i)
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
cursor.execute('SELECT "COMMUNITY_AREA_NAME" C, "PERCENT_HOUSEHOLDS_BELOW_POVERTY" P FROM public."chicago_census_data" ORDER BY P DESC LIMIT 5')

for i in cursor.fetchall():
    print(i)
```

### Problem 8

##### Which community area is most crime prone?


```python
cursor.execute('SELECT "COMMUNITY_AREA_NUMBER" C, COUNT(*) S FROM public."chicago_crime_data" WHERE "COMMUNITY_AREA_NUMBER" IS NOT NULL GROUP BY C ORDER BY S DESC LIMIT 1')

for i in cursor.fetchall():
    print(i)
```
