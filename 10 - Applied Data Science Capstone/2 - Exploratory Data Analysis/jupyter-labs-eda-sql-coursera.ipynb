## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
result = c.execute('SELECT DISTINCT Launch_Site from spacex')

for r in result:
    print(r[0])
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
result = c.execute('SELECT * FROM spacex WHERE Launch_Site LIKE "CCA%" LIMIT 5')

for r in result:
    print(r)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
result = c.execute(
    'SELECT SUM(PAYLOAD_MASS__KG_) FROM spacex WHERE Customer = "NASA (CRS)"')

for r in result:
    print(r)
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
result = c.execute(
    'SELECT AVG(PAYLOAD_MASS__KG_) FROM spacex WHERE Booster_Version = "F9 v1.1"')

for r in result:
    print(r)
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
result = c.execute(
    'SELECT MIN(Date) FROM spacex WHERE "Landing _Outcome" = "Success"')

for r in result:
    print(r)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
result = c.execute(
    'SELECT Booster_Version FROM spacex WHERE "Landing _Outcome" = "Success (drone ship)" AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000')

for r in result:
    print(r)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
result = c.execute(
    'SELECT "Landing _Outcome", COUNT("Landing _Outcome") FROM spacex GROUP BY "Landing _Outcome" ORDER BY "Landing _Outcome"')

for r in result:
    print(r)
```

### Task 8

##### List the names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
result = c.execute(
    '''
    SELECT Booster_Version, PAYLOAD_MASS__KG_ FROM spacex
    WHERE 
        PAYLOAD_MASS__KG_ = 
        (
            SELECT MAX(PAYLOAD_MASS__KG_) FROM spacex)'''
        )

for r in result:
    print(r)
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
# for some reason couldn't extract year from Date column so i used string ends with 2015 instead

result = c.execute(
    '''
    SELECT "Landing _Outcome", Booster_Version, Launch_Site, Date 
    FROM spacex 
    WHERE "Landing _Outcome" = "Failure (drone ship)" 
    AND Date LIKE "%2015"
    ''')

for r in result:
    print(r)


```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
result = c.execute(
    '''
    SELECT "Landing _Outcome", COUNT("Landing _Outcome") as c, Date
    FROM spacex 
    WHERE Date >= '04-06-2010' AND Date <= '20-03-2017'
    GROUP BY "Landing _Outcome" 
    ORDER BY "c"
    ''')

for r in result:
    print(r)
```
